**Inicialização da função para converter de <span style="color:orange">Binário</span> para <span style="color:orange">Decimal</span>**

```python
def BinarioDecimal(valor):
    dec = 0
    pot = 0
    for i in range(len(valor) - 1, -1, -1):
        dec = dec + pow(2, pot)*int((valor[i]))
        pot = pot + 1
    return dec
```

**Inicialização da função para converter de <span style="color:orange">Decimal</span> para <span style="color:orange">Binário</span>**

```python
def DecimalBinario (d):
    resposta =""
    b = ""
    if d < 0:
        return("Número inválido, pois é negativo!")
    elif d  == 0:
        b = "00000"
        return b
    elif d == 1:
        b = "00001"
        return b
    elif d  == 2:
        b = "00010"
        return b
    elif d == 3:
        b = "00011"
        return b
    elif d >= 4:
        while d != 1:
            b = b+str(d%2)
            d = d//2
            if d//2 == 1:
                b = b+str(d%2)+str(d//2)
                break
        for i in range(len(b)-1, - 1, -1):
            resposta += b[i]
        while len(resposta) < 5:
            resposta = "0" + resposta
    return resposta
```

**Inicialização da função para converter de <span style="color:orange">Binário</span> para <span style="color:orange">Hexadecimal</span>**

```python
def binHex(x):

    while((len(x))%4)!= 0:
        x = "0" + x

    h = ""
    for i in range(0, len(x), 4):
        if(x[i] + x[i+1] + x[i+2] + x[i+3] == "0000"):
            h = h + "0"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "0001"):
            h = h + "1"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "0010"):
            h = h + "2"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "0011"):
            h = h + "3"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "0100"):
            h = h + "4"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "0101"):
            h = h + "5"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "0110"):
            h = h + "6"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "0111"):
            h = h + "7"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "1000"):
            h = h + "8"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "1001"):
            h = h + "9"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "1010"):
            h = h + "A"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "1011"):
            h = h + "B"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "1100"):
            h = h + "C"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "1101"):
            h = h + "D"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "1110"):
            h = h + "E"
        elif(x[i] + x[i+1] + x[i+2] + x[i+3] == "1111"):
            h = h + "F"
    return h
```

**Inicialização da função para converter de <span style="color:orange">Hexadecimal</span> para <span style="color:orange">Binário</span>**

```python
def HexParaBinario(n):
    b = ""
    for i in range(len(n)):
        if n[i] == "0":
            b = b + "0000"
        elif n[i] == "1":
            b = b + "0001"
        elif n[i] == "2":
            b = b + "0010"
        elif n[i] == "3":
            b = b + "0011"
        elif n[i] == "4":
            b = b + "0100"
        elif n[i] == "5":
            b = b + "0101"
        elif n[i] == "6":
            b = b + "0110"
        elif n[i] == "7":
            b = b + "0111"
        elif n[i] == "8":
            b = b + "1000"
        elif n[i] == "9":
            b = b + "1001"
        elif n[i] == "A":
            b = b + "1010"
        elif n[i] == "B":
            b = b + "1011"
        elif n[i] == "C":
            b = b + "1100"
        elif n[i] == "D":
            b = b + "1101"
        elif n[i] == "E":
            b = b + "1110"
        elif n[i] == "F":
            b = b + "1111"
        else:
            return False
    return b
```

**Função para <span style="color:green">QUEBRAR</span>, o <span style="color:orange">ENDERECO</span>, ou seja, valores dos seus Bytes serão <span style="color:orange">PECORRIDOS</span> e será feita a devida quebra em <span style="color:green">Tag</span>, <span style="color:green">Byte</span>, <span style="color:green">Conj</span>.**

```python
def QuebraEndereco(endereco):
    global Tag
    global Conj
    global Byte
    global Bloco
    Tag = ""
    Conj = ""
    Byte = ""
    Bloco = ""
    tamanhoEndereco = len(endereco)
    for i in range(len(endereco)):
        if i < TamanhoTag:
            Tag = Tag + endereco[i]
        elif i < TamanhoTag + TamanhoConjunto:
            Conj = Conj + endereco[i]
        elif i < TamanhoTag + TamanhoConjunto + TamanhoByte:
            Byte = Byte + endereco[i]
    print("TAG:", Tag)
    print("CONJUNTO:", Conj)
    print("BYTE:", Byte)
    return ""
```

**Função para <span style="color:green">ENCONTRAR</span>, o <span style="color:orange">DADO</span>, correspondente ao endereço informado pelo usuário, em que se procura específicamente na <span style="color:orange">TAG</span>, caso haja tal <span style="color:magenta">TAG</span> na <span style="color:magenta">CACHE</span> é dado um <span style="color:green">HIT</span>, caso contrário, <span style="color:red">MISS</span> e começa a procura de acordo com o endereço na <span style="color:red">MP</span>. Uma <span style="color:green">LINHA</span> aleatória é criada para receber o bloco da MP, cujo corresponde ao endereço.**

```python
def EncontraDadoCache():
    global linha
    global indice
    global Bloco
    procura = (BinarioDecimal(Conj)) * LinhaConjunto
    busca = False
    for i in range((procura), (procura + 2)):
        if Tag == Cache[i][0]:
            indice = i
            busca = True
            return busca
            break
    if busca == False:
        Bloco = int(BinarioDecimal(Tag + Conj))
        linha = random.randrange((procura), (procura + 2))
        Cache[linha][0] = Tag
        Cache[linha][2] = MP[(Bloco * 2)]
        Cache[linha][1] = MP[(Bloco * 2) + 1]
        busca = False
    return busca
```

**Função para <span style="color:green">SOMAR</span>, dois <span style="color:orange">Numéros</span> <span style="color:green">Binários</span> bit a bit**

```python
def somarBinario(b1, b2):
    global resto
    if resto == "1":
        if (b1 == "1" and b2 == "0") or (b1 == "0" and b2 == "1"):
            resto = "1"
            return "0"
        elif b1 == "0" and b2 == "0":
            resto = "0"
            return "1"
        elif b1 == "1" and b2 == "1":
            resto = "1"
            return "1"
    else:
        if (b1 == "1" and b2 == "0") or (b1 == "0" and b2 == "1"):
            return "1"
        elif b1 == "0" and b2 == "0":
            return "0"
        elif b1 == "1" and b2 == "1":
            resto = "1"
            return "0"
```

**Função para <span style="color:green">SUBTRAIR</span>, dois <span style="color:orange">Numéros</span> <span style="color:green">Binários</span> usando complemento de 1 e 2.**

```python
def subtrairBinario(b1, b2):
    print("Execução:" + b1 + " - " + b2)
    C1 = ""
    for i in range(len(b2)):
        if b2[i] == "0":
            C1 += "1"
        else:
            C1 += "0"
    print("\nBinário: " + b2 + "\nComplemento de 1: " + C1)

    u = "1"
    while len(C1) > len(u):
        u = "0" + u

    global resto
    resto = "0"
    C2 = ""
    for i in range(len(C1) - 1, -1, -1):
        C2 = somarBinario(C1[i], u[i]) + C2
    print("Complemento de 2: " + C1 + " + 1 = " + C2)

    resto = "0"
    res = ""
    for i in range(len(C2) - 1, -1, -1):
        res = somarBinario(b1[i], C2[i]) + res
    print("\nResultado final: \n" + b1 + " + " + C2 + " = " + str(res))
    return res
```

**Função para <span style="color:green">SEPARAR</span> o <span style="color:orange">Operando</span>, ou seja, o <span style="color:green">Dado</span> que será usado no programa. Os bits restantes do dado completo excluindo os 5 primeiros (COP)**

```python
def separaOperando(dado):
    global Operando
    Operando = ""
    for i in range(len(dado)):
        if(i > 6):
            Operando = Operando + dado[i]
        if(i == 11):
            break
```

**Função para <span style="color:green">DETERMINAR</span> o <span style="color:orange">COP</span>, ou seja, a <span style="color:green">Instrução</span> de acordo com os 5 bits que foram separados usando a função acima.**

```python
def determinaCOP(cop):
    global R0
    global RI
    global REM
    global RDM
    global CI
    global Operando
    global Indicador
    print("\nOperação ", binHex(cop))
    if(cop == "0001"): #Load
        print("LDA")
        separaOperando(RI)
        REM = Operando
        print("REM recebe endereço do operando: "+ REM)
        RDM = ativarLeitura(REM)
        print("RDM recebe dado do operando: "+ RDM)
        R0 = RDM
        return ("R0 recebe: ", R0)
    elif (cop == "0010"): #STR
        print("STR")
        separaOperando(RI)
        REM = Operando
        print("REM recebe endereço do operando: "+ REM)
        RDM = R0
        print("RDM recebe dado do R0: "+ RDM)
        ativarEscrita(REM, RDM)
        return ("Dado inserido na MP")
    elif (cop == "0011"):#ADD
        global sobra
        global soma
        soma = ""
        sobra = ""
        separaOperando(RI)
        REM = Operando
        print("REM recebe endereço do operando: "+ REM)
        RDM = ativarLeitura(REM)
        print("RDM recebe dado do operando: "+ RDM)
        for i in range(3):
            soma = somarBinario(R0[i], RDM[i]) + soma
        R0 = soma
        return ("Soma resultante: R0 recebe "+ R0)
    elif (cop == "0100"):#SUB
        print("SUB")
        Indicador = 0
        separaOperando(RI)
        REM = Operando
        print("REM recebe endereço do operando: "+ REM)
        RDM = ativarLeitura(REM)
        print("RDM recebe dado do operando: "+ RDM)
        if(BinarioDecimal(R0) > BinarioDecimal(RDM)):
            Indicador = 0
        elif(BinarioDecimal(R0) < BinarioDecimal(RDM)): 
            Indicador = 1
        R0 = subtrairBinario(R0, RDM)
        return ("Subtração resultante: R0 recebe " + R0)
    elif (cop == "0101"):#JZ
        print("JZ")
        if(BinarioDecimal(R0) == 0):
            separaOperando(RI)
            CI = Operando
            print("Jump realizado com sucesso. CI:")
            return CI
        else:
            return ("Jump falhou")
    elif (cop == "0110"):#JP
        print("JP")
        if(Indicador == 0):
            separaOperando(RI)
            CI = Operando
            print("Jump realizado com sucesso. CI:")
            return CI
        else:
            return ("Jump falhou")
    elif (cop == "0111"):#JN
        print("JN")
        if(Indicador == 1):
            separaOperando(RI)
            CI = Operando
            print("Jump realizado com sucesso. CI:")
            return CI
        else:
            return ("Jump falhou")
    elif (cop == "1000"):#JMP
        print("JMP")
        separaOperando(RI)
        CI = Operando
        print("Jump realizado com sucesso. CI:")
        return CI
    elif (cop == "1001"):#GET
        print("GET")
        dado = input("Digite um dado de 12 bits:")
        separaOperando(RI)
        REM = Operando
        print("REM recebe endereço do operando: " + REM)
        RDM = dado
        print("RDM recebe dado do operando: " + RDM)
        ativarEscrita(REM, RDM)
        return ("Dado inserido na MP")
    elif (cop == "1010"):#PRT
        print("PRT")
        separaOperando(RI)
        REM = Operando
        print("REM recebe endereço do operando: "+ REM)
        RDM = ativarLeitura(REM)
        print("RDM recebe dado do operando: "+ RDM)
        return ("Retorno: "+RDM+" - "+binHex(RDM))
```

**Função para <span style="color:green">EXECUTAR</span> o <span style="color:orange">Programa</span>, ou seja, as <span style="color:green">Instruções</span> serão determinadas e farão suas respectivas funções. Aqui que o CI será usado para apontar para os próximos endereços/instruções. Os ciclos serão incrementados aqui também.**

```python
def executarPrograma():
    global CI
    global COp
    global RI
    global REM
    global RDM
    RDM = ""
    CI = "00000"
    ciclo = 0
    COp = "" #Teste
    while True:
        ciclo = ciclo + 1
        print("\nCiclo:", ciclo)
        REM = CI
        print("REM", REM, " - ", binHex(REM))
        CI = DecimalBinario(BinarioDecimal(CI) + 1)
        print("Instrução do próximo ciclo: CI = ", CI, " - ", binHex(CI))
        RDM = ativarLeitura(REM)
        print("RDM", RDM, " - ", binHex(RDM))
        RI = RDM
        print("RI", RI," - ", binHex(RI))
        serparaCOP(RI)
        if(COp == "0000"):
            print("Operação:", binHex(COp),"\nHLT")
            break
        else:
            print(determinaCOP(COp))
```
