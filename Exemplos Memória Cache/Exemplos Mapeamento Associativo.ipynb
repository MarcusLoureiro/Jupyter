#### O Mapeamento Associativo da memória cache se baseia na técnica em que cada bloco da MP é flexível em relação a linha da Cache, buscando resolver os problemas do Direto. O programa a seguir servirá exclusivamente para esse tipo de mapeamento, portanto é preciso ter cuidado ao inserir os dados que serão pedidos, pois pode alterar todo o resultado.

```python
import math

aviso = """ERRO - DADO INCORRETO, TENTE NOVAMENTE:"""
pergunta = """Digite uma das opções a seguir:
1 - Calcular quantidade de bits necessários para a Memória Cache
2 - Calcular o formato de endereço da Memória Cache.
3 - Calcular o valor binário do campo bloco localizado.
0 - Encerrar programa
Digite aqui:"""

Escolha = int(input(pergunta))
```

#### A parte do código acima irá servir para iniciar as variáveis de ''aviso'' e "pergunta". Implementar a biblioteca math, pois usaremos metódos dela para facilitar e tornar o código mais eficiente e compacto. Por fim irá se perguntar a Escolha do usuário, de acordo com uma das opções, 1, 2, 3 ou 0.

```python
def HexParaBinario(n):
    b = ""
    for i in range(len(n)):
        if n[i] == "0":
            b = b + "0000"
        elif n[i] == "1":
            b = b + "0001"
        elif n[i] == "2":
            b = b + "0010"
        elif n[i] == "3":
            b = b + "0011"
        elif n[i] == "4":
            b = b + "0100"
        elif n[i] == "5":
            b = b + "0101"
        elif n[i] == "6":
            b = b + "0110"
        elif n[i] == "7":
            b = b + "0111"
        elif n[i] == "8":
            b = b + "1000"
        elif n[i] == "9":
            b = b + "1001"
        elif n[i] == "A":
            b = b + "1010"
        elif n[i] == "B":
            b = b + "1011"
        elif n[i] == "C":
            b = b + "1100"
        elif n[i] == "D":
            b = b + "1101"
        elif n[i] == "E":
            b = b + "1110"
        elif n[i] == "F":
            b = b + "1111"
        else:
            return False
    return b
```

#### Usaremos para quando usário  escolher a opção 3, um conversor HEX para BIN, veja que se não for os carcateres de um HEX, um valor False é retornado.

```python
def ValidaEntrada(b):
    for i in range(len(b)):
        if b[i] == "B":
            return True
            break
        elif b[i] + b[i + 1] == "KB":
            return True
            break
        elif b[i] + b[i + 1] == "MB":
            return True
            break
        elif b[i] + b[i + 1] == "GB":
            return True
            break
        elif b[-1] != "B":
            return False
```

#### A função acima vai validar a unidade das entradas se está em KB, MB, GB, B e etc.

```python
def CalculaBase(b):
    valor = ""
    resultado = 1
    for i in range(len(b)):
        if b[i] == "B":
            break
        elif b[i] == "K":
            resultado = 2 ** 10
            break
        elif b[i] == "M":
            resultado = 2 ** 20
            break
        elif b[i] == "G":
            resultado = 2 ** 30
            break
        else:
            valor = valor + b[i]

    return (int(valor)) * resultado
```

#### Como já validamos a entrada, outra função irá ser chamada, essa irá calcular de acordo como a unidade a base e o expoente correspondete, sempre é melhor trabalhar com base 2 quando manipulamos dados computacionais, pois é assim que a máquina entende.

```python
def InverteBase(b):
    if (b % (2 ** 30)) == 0:
        return str(b / (2 ** 30)) + "Gb"
    elif (b % (2 ** 20)) == 0:
        return str(b / (2 ** 20)) + "Mb"
    elif (b % (2 ** 10)) == 0:
        return str(b / (2 ** 10)) + "Kb"
    else:
        return str(b) + "B"
```

#### Essa função serve apenas para inverter a base que calculamos para valores mais compreensíveis para o usuário.

```python
def CalculaResultado(A, B, C, Escolha):
    if Escolha == 1:
        """Total de bits de dados"""
        Tdados = A * 8
        """Quantidade de linhas"""
        Linhas = A / B
        """Quantidade de blocos"""
        Blocos = C / B
        """Largura do Campo Bloco"""
        Bloco = math.log2(Blocos)
        """Total de bits do campo banco"""
        TBloco = Linhas * Bloco
        """Total de bits da cache"""
        Tcache = Tdados + TBloco
        print("Bits necessários para a Cache em questão:", InverteBase(Tcache))

    elif Escolha == 2:
        """Total de Blocos"""
        Tblocos = A / B
        """Bits da Tag"""
        Bloco = math.log2(Tblocos)
        """Bits da Célula"""
        celula = math.log2(B)
        print(
            "\nFormato de endereço da cache:\n| Bloco |",
            Bloco,
            " bits\n| Byte  |",
            celula,
            " bits\nTotal de bits do endereço: ",
            (Bloco + celula),
            " bits\n" + "Número de linhas da cache: ",
            (B),
        )

    elif Escolha == 3:
        """Bits da celula"""
        celula = math.log2(A)
        """Quantidade de blocos"""
        Blocos = (2 ** (len(C) * 4)) / A
        """Bits do campo bloco"""
        Bloco = math.log2(Blocos)
        """Endereço em binário"""
        Abin = HexParaBinario(C)
        Abloco = ""
        Acelula = ""
        for i in range(len(Abin)):
            if i >= 0 and i < Bloco:
                Abloco = Abloco + Abin[i]
            elif i >= Bloco:
                Acelula = Acelula + Abin[i]
        print(
            "\nFormato de endereço da cache:\n| Bloco |",
            Abloco,
            "\n| Byte  |",
            Acelula,
            "\nEndereço do bloco para acesso: ",
            Abloco,
        )
```

#### Essa pode-se dizer como a parte mais importante do código, nela, calcula-se todos os resultados, de acordo com a escolha do  usuário, e exibi-se uma resposta por vez. A última seção do código está embaixo, nela serão chamadas todas as funções acima e portanto se validará cada entrada do usuário e por fim se exibe o resultado final, e pedirá uma nova escolha, em que se encerra o programa somente ao digitar na hora da escolha, o digito 0.

```python
while True:
    if Escolha == 1:
        Ccache = input(
            """Digite a capacidade da memóra cache:
Ex: 32KB
Digite aqui:"""
        )
        while True:
            if ValidaEntrada(Ccache):
                Ccache = CalculaBase(Ccache)
                break
            else:
                print(aviso)
                Ccache = input(
                    """Digite a capacidade da memóra cache:
Ex: 32KB
Digite aqui:"""
                )
        Nb = input(
            """Digite o número de células por linha da memória cache:
Ex: 8B
Digite aqui:"""
        )
        while True:
            if ValidaEntrada(Nb):
                Nb = CalculaBase(Nb)
                break
            else:
                print(aviso)
                Nb = input(
                    """Digite o número de células por linha da memória cache:
Ex: 8B
Digite aqui:"""
                )
        Cmp = input(
            """Informe a capacidade de células(bytes) da memória principal:
Ex: 16MB
Digite aqui:"""
        )
        while True:
            if ValidaEntrada(Cmp):
                Cmp = CalculaBase(Cmp)
                break
            else:
                print(aviso)
                Cmp = input(
                    """Informe a capacidade de células(bytes) da memória principal:
Ex: 16MB
Digite aqui:"""
                )
        print(CalculaResultado(Ccache, Nb, Cmp, Escolha))
        Escolha = int(input(pergunta))

    elif Escolha == 2:
        Cmp = input(
            """Informe a capacidade de células(bytes) da memória principal:
Ex: 16MB
Digite aqui:"""
        )
        while True:
            if ValidaEntrada(Cmp):
                Cmp = CalculaBase(Cmp)
                break
            else:
                print(aviso)
                Cmp = input(
                    """Informe a capacidade de células(bytes) da memória principal:
Ex: 16MB
Digite aqui:"""
                )

        Nb = input(
            """Digite o número de células por linha da memória cache:
Ex: 8B
Digite aqui:"""
        )
        while True:
            if ValidaEntrada(Nb):
                Nb = CalculaBase(Nb)
                break
            else:
                print(aviso)
                Nb = input(
                    """Digite o número de células por linha da memória cache:
Ex: 8B
Digite aqui:"""
                )
        Lb = input(
            """Informe o total de linhas da memória cache:
Ex: 4K
Digite aqui:"""
        )
        while True:
            if ValidaEntrada(Lb):
                Lb = CalculaBase(Lb)
                break
            else:
                print(aviso)
                Lb = input(
                    """Informe o total de linhas da memória cache:
Ex: 4K
Digite aqui:"""
                )
        print(CalculaResultado(Cmp, Nb, Lb, Escolha))
        Escolha = int(input(pergunta))

    elif Escolha == 3:
        Lblocos = input(
            """Digite a largura dos blocos da memóra principal:
Ex: 16B
Digite aqui:"""
        )
        while True:
            if ValidaEntrada(Lblocos):
                Lblocos = CalculaBase(Lblocos)
            break
        else:
            print(aviso)
            Lblocos = input(
                """Digite a largura dos blocos da memóra principal:
Ex: 16B
Digite aqui:"""
            )
        NCache = input(
            """Digite o número de células da memória cache:
Ex: 64KB
Digite aqui:"""
        )
        while True:
            if ValidaEntrada(NCache):
                NCache = CalculaBase(NCache)
                break
            else:
                print(aviso)
                NCache = input(
                    """Digite o número de células da memória cache:
Ex: 64KB
Digite aqui:"""
                )
        Endereco = input(
            """Informe o endereço a ser acessado em Hexadecimal:
Ex: 2A56F
Digite aqui:"""
        )
        while True:
            if HexParaBinario(Endereco) == False:
                print(aviso)
                Endereco = input(
                    """Informe o endereço a ser acessado em Hexadecimal:
Ex: 2A56F
Digite aqui:"""
                )
            else:
                break
        print(CalculaResultado(Lblocos, NCache, Endereco, Escolha))
        Escolha = int(input(pergunta))
    elif Escolha == 0:
        print("PROGRAMA ENCERRADO")
        break
```
